### Step 1. Define study functions

```python
# Visualizing histogram, probability plot and box-plot

def diagnostic_plots(data, variable):
    fig = plt.figure()
    plt.figure(figsize = (18, 5))
   
    # Histogram
    plt.subplot(1, 3, 1)
    fig = data[variable].hist(bins = 30)
    fig.set_ylabel ('Number of passengers')
    fig.set_xlabel(variable)
    
    # Probability plot
    plt.subplot(1, 3, 2)
    stats.probplot(data[variable], dist = 'norm', plot = plt)
    
    # Box-plot
    plt.subplot(1, 3, 3)
    fig = data.boxplot(column = variable)
    fig.set_title('Box-plot')
    fig.set_ylabel(variable)
    
    plt.tight_layout()
    plt.show()
```

```python
# Show in DataFrame missing values with type of variable

def missing_value_info(data):
    nan_list = [var for var in data.columns if data[var].isnull().sum() > 0]
    info_na = data[nan_list].isnull().mean().round(3).sort_values(ascending = False).reset_index()
    info_na.columns = ['Columns','Nan_mean']
    info_na['Type'] = info_na.Columns.map(data.dtypes)
    info_na.Type = np.where(info_na.Type == 'O','Categorical', info_na.Type)
    info_na.Type = np.where(info_na.Type != 'Categorical','Numerical', info_na.Type)

    return info_na

# Visualizing in plot missing values
def missing_value_info_plot(data):    
    info = missing_value_info(data)
    plt.figure(figsize=(len(info.Columns) * 1.5, 2.5))
    plt.plot(info.Nan_mean,'b.-')
    plt.axhline(y = 0.5,color = 'red')
    plt.yticks(np.arange(0, 1.2, 0.2))
    plt.xticks(info.Columns.index,info.Columns,rotation = 90)
    plt.ylabel('Percentage of missing data')
    plt.grid()
    
    plt.show()
```

### Logistic regression by MeanMedian imputation

```python
X_test.isnull().sum()
```

```python
# Create Pipeline

Logit_mean_pipe = Pipeline(steps = [
    ('imputer_num', MeanMedianImputer(imputation_method = 'mean', variables = ['age'])),
    ('encoder_rare_label', ce.RareLabelEncoder(tol = 0.05, n_categories = 2, variables = cat_list)),
    ('encoder', ce.CountFrequencyEncoder(encoding_method = 'count', variables = cat_list)),
    ('discritizator', EqualWidthDiscretiser(bins = 8, variables = ['age', 'fare'])),
    #('capping', Winsorizer(capping_method = 'quantiles', tail = 'both', fold = 0.05)),
    ('constant', DropConstantFeatures(tol = 0.998)),
    ('duplicated', DropDuplicateFeatures()),
    ('smart_corr', DropCorrelatedFeatures()),
    ('scaler', StandardScaler()),
    ('logit', LogisticRegression(tol = 0.5, max_iter = 1000))
])
```

```python
# Create parameters for GridSearch

Logit_mean_param_grid = {
    'imputer_num__imputation_method': ['mean', 'median'],
    'encoder_rare_label__tol': [0.04, 0.05, 0.06, 0.07],
    'encoder_rare_label__n_categories': [2, 3, 4, 5, 6, 7],
    'encoder': [
        ce.OneHotEncoder(),
        ce.OrdinalEncoder(),
        ce.CountFrequencyEncoder()
    ],
    'discritizator__bins': [5, 6, 7, 8, 9, 10],
    #'capping__tail': ['both', 'left', 'right'],
    'scaler': [MinMaxScaler(),
              StandardScaler(),
              RobustScaler()],
    'logit__max_iter': [500, 1000, 2000, 3000],
    'logit__C': [1.0, 1.2, 1.3, 1.5]
    
}
```

```python
# Create GridSearch

Logit_mean_grid_search = GridSearchCV(estimator = Logit_mean_pipe, 
                                 param_grid = Logit_mean_param_grid, 
                                 cv = 5, 
                                 n_jobs = -1, 
                                 scoring = 'accuracy', 
                                 error_score = 'raise')
```

```python
for i in tqdm(Logit_mean_pipe):
    Logit_mean_grid_search.fit(X_train, y_train)
```
